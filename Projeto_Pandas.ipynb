```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = pd.read_csv('/content/drive/MyDrive/Datasets/gastos.csv', parse_dates=['data'])
```

```python
df.head()
```

```python
print("Total de entradas", df[df['tipo'] == 'Entrada']['valor'].sum())
print("Total de saídas", df[df['tipo'] == 'Saída']['valor'].sum())
print("Saldo final:", df['valor'].sum())
```

```python
totais = df.groupby('tipo')['valor'].sum()
totais.plot(kind='bar', color=['green', 'red'], title='Entradas vs Saídas')
plt.ylabel('R$')
plt.show()
```

```python
gastos = df[df['tipo'] == 'Saída'].groupby('descricao')['valor'].sum().sort_values()
gastos.plot(kind='barh', title='Gastos por Categoria', color='orange')
plt.xlabel('Valor (negativo)')
plt.show()
```

```python
df = df.sort_values('data')
df['saldo_acumulado'] = df['valor'].cumsum()

plt.figure(figsize=(10,5))
plt.plot(df['data'], df['saldo_acumulado'], marker='o')
plt.title('Evolução do Saldo')
plt.xlabel('Data')
plt.ylabel('Saldo (R$)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
